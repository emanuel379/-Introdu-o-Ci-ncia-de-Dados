**3** - Conside os dados da Iris e mostre a distribuição de probabilidades de cada uma das variáveis após a normalização e padronização.

```python
df3 = pd.read_csv('data/iris.csv')
df3
```

```python
# Normalização

from sklearn.preprocessing import MinMaxScaler
# considera somente os atributos, ignorando a última coluna, que contem a classe
X = np.array(df3[df3.columns[0:df3.shape[1]-1]])
for i  in range(X.shape[1]):
    print("Maior valor da coluna:", i, ":", max(X[:,i]))
    print("Menor Valor da coluna:", i, ":", min(X[:,i]))
# prepara a função para transformar os dados
scaler = MinMaxScaler(feature_range=(0, 1))
# Realiza a normalização e coloca em um novo vetor
X_norm = scaler.fit_transform(X)
print(X_norm)
```

```python
# Distribuições de probabilidade
import matplotlib.pyplot as plt
# sepal_length

DfX_norm = pd.DataFrame(X_norm)
x = np.array([DfX_norm.columns[0]])

plt.figure(figsize=(10, 10))
plt.title('normalized sepal_length')
plt.hist(x)
plt.show()
```

```python
DfX_norm = pd.DataFrame(X_norm)
x = np.array([DfX_norm.columns[1]])

plt.figure(figsize=(10, 10))
plt.title('normalized sepal_width')
plt.hist(x)
plt.show()
```

```python
DfX_norm = pd.DataFrame(X_norm)
x = np.array([DfX_norm.columns[2]])

plt.figure(figsize=(10, 10))
plt.title('normalized petal_length')
plt.hist(x)
plt.show()
```

```python
DfX_norm = pd.DataFrame(X_norm)
x = np.array([DfX_norm.columns[3]])

plt.figure(figsize=(10, 10))
plt.title('normalized petal_width')
plt.hist(x)
plt.show()
```

```python
# Padronização

from sklearn.preprocessing import StandardScaler
Y = np.array(df3[df3.columns[0:df3.shape[1]-1]])
# prepara a função para transformar os dados
scaler = StandardScaler().fit(Y)
# realiza a padronização (média=0, variância = 1)
rescaledY = scaler.transform(Y)
print(rescaledY)
```

```python
# Distribuição de probabilidade

DfrescaledY = pd.DataFrame(rescaledY)
y = np.array(DfrescaledY[DfrescaledY.columns[0]])

plt.figure(figsize=(10, 10))
plt.title('standardized sepal_length')
plt.hist(y)
plt.show
```

```python
DfrescaledY = pd.DataFrame(rescaledY)
y = np.array(DfrescaledY[DfrescaledY.columns[1]])

plt.figure(figsize=(10, 10))
plt.title('standardized sepal_width')
plt.hist(y)
plt.show
```

```python
DfrescaledY = pd.DataFrame(rescaledY)
y = np.array(DfrescaledY[DfrescaledY.columns[2]])

plt.figure(figsize=(10, 10))
plt.title('standardized petal_length')
plt.hist(y)
plt.show
```

```python
DfrescaledY = pd.DataFrame(rescaledY)
y = np.array(DfrescaledY[DfrescaledY.columns[3]])

plt.figure(figsize=(10, 10))
plt.title('standardized petal_width')
plt.hist(y)
plt.show
```

**6** - Encontre os outliers nos dados das bases Iris e BostonHouse.

```python
df6 = pd.read_csv('data/iris.csv')
df6
```

```python
import seaborn as sns
sns.pairplot(df6)
plt.show()
```

```python
Q1 = df6.quantile(0.25)
Q3 = df6.quantile(0.75)
IQR = Q3 - Q1

print (((df6 < (Q1 - 1.5 * IQR)) | (df6 > (Q3 + 1.5 * IQR))))
```

```python
# Observando linhas onde existem outliers

print(((df6 < (Q1 - 1.5 * IQR)) | (df6 > (Q3 + 1.5 * IQR))).any(axis=1)) # Onde axis = 1 representa a coluna
```
